```python
#Reading the file and converting it as a dict. 

import pandas as pd
import xmltodict
```

```python
#Reading the file and converting it as a dict. 
input_path = r"C:\\Users\\Yoselyn\\Documents\\HEALTH APP EXPORT\\export\\apple_health_export\\export.xml"
with open(input_path, 'r') as xml_file:
    input_data = xmltodict.parse(xml_file.read())
```
